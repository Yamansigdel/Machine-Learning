```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data_student=pd.read_csv('./Student_scores.csv')
display(data_student.head(6))
```

<center><h1>Least Square method</h1></center>

```python
x=data_student['Hours'].tolist()
y=data_student['Scores'].tolist()
plt.scatter(x,y)
plt.xlabel('hours')
plt.ylabel('scores')


x_tot=0
y_tot=0
count=0
for i in x:
    x_tot+=i
    count+=1

for i in y:
    y_tot+=i

x_mean=x_tot/count
y_mean=y_tot/count

m=0
num=0 
den=0
for i in range(count):
    num+=(x[i]-x_mean)*(y[i]-y_mean)
    den+=(x[i]-x_mean)**2

m=num/den

c=y_mean - (m * x_mean)
print(m,c)


#y_pred=[]
# for i in x:
#     Y_p=m * i+ c
#     y_pred.append(Y_p)
y_pred=[m * i + c for i in x]

plt.plot(x,y_pred,color='red')

#For error visualization, calculate R_sqr
#It says how well the best fit line justifies the data
R_num=0
R_den=0
for i in range(count):
    R_num+=(y[i]-y_pred[i])**2
    R_den+=(y[i]-y_mean)**2


R_sqr= 1 - (R_num/R_den)
print(R_sqr)

x_inf=float(input('Enter hrs:'))
y_inf=m * x_inf + c
print(x_inf, y_inf)

```

<center><h1>Iterative method</h1></center>

```python

m= 5 
c=5 
alpha=0.01
print(f'm : {m}     c: {c}')
# d_cost_m=2* sum_m 
# d_cost_c=2* sum_c

for i in range(10000):

    sum_m=0
    sum_c=0
    for i in range(count):

        y_pred=m * x[i] + c
        
        sum_m+=(y_pred-y[i])* x[i]
        sum_c+=(y_pred-y[i])
    
    sum_m=sum_m * (2/count)
    sum_c=sum_c * (2/count)

    m= m - ( alpha * sum_m)
    c= c - ( alpha * sum_c)
    print(f'm : {m}     c: {c}')

R_num=0
R_den=0
y_pred=[m * i + c for i in x]
for i in range(count):
    
    R_num+=(y[i]-y_pred[i])**2
    R_den+=(y[i]-y_mean)**2


R_sqr= 1 - (R_num/R_den)
print(R_sqr)

# print(f'Final m: {m}     Final c: {c}')
```
