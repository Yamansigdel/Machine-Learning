<h3>Tensorflow Development</h3>

```python
import pandas as pd 
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Flower_dataset.csv')
features=np.array(df[['X1','X2']])
labels=np.array(df['Y']).reshape(-1,1)
# print(features,labels)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,models
```

```python

model=models.Sequential()
model.add(layers.Dense(64,activation='relu',input_dim=(df.drop(columns=['Y'])).shape[1]))
model.add(layers.Dense(128,activation='relu'))
model.add(layers.Dense(128*2,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))

#choose loss according to output values.
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics='accuracy')

X=np.array(df.drop(columns=['Y']))
Y=np.array(df['Y']).reshape(-1,1)
model.fit(X,Y,epochs=500)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
model.fit(X_train,y_train,epochs=300,validation_data=(X_test,y_test))
```

```python
# model.evaluate(X_test) #eslai evaluate gareko bhancha
a=np.array([[3.0,0.4]])
# a=np.array([[3.0,0.4]]) #yo chai 24% ma paryo , 0 aaunuparne 1 aaucha so if output[0][0] < 0.54: gareko aafai accuracy milauna
output=(model.predict(a))  #eslai inference gareko bhancha
print(output)
if output[0][0] < 0.54:
  output=0
else:
  output = 1

print(output)
```

```python
features= np.array(data[['X1','X2','X3','X4']])
labels=np.array(data['Y']).reshape(-1,1)
display(features,labels)
```

```python
model=models.Sequential()
model.add(layers.Dense(64,activation='relu',input_dim=4))
model.add(layers.Dense(128,activation='relu'))
model.add(layers.Dense(128*2,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])

X_train,X_test,Y_train,Y_test = train_test_split(features,labels,test_size=0.1,random_state=42)

X_train = np.array(data.drop(columns=['Y']))
Y_train = np.array(data['Y']).reshape(-1,1)
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=300)
```

```python
a=np.array([[1,1,0,1]])
output=(model.predict(a))  
print(output)
if output[0][0] < 0.5:
  output=0
else:
  output = 1

print(output)
```
